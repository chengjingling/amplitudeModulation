This exercise performs amplitude demodulation to bring the audio files to an audible range of frequencies, and plays the audio files to determine which one contains the secret code.

Firstly, we need to import the libraries required in the application.

```python
# Import libraries
import numpy as np
import scipy.io.wavfile
from IPython.display import Audio
```

Next, we have the amplitude_demodulation function that performs amplitude demodulation on a signal.

The function takes in a signal and its sample rate as the parameters, and generates time values based on the length of the input signal and the sample rate. A carrier signal is then generated at a frequency of 18000 Hz, and multiplied with the signal to get the demodulated signal.

```python
# Function to perform amplitude demodulation
def amplitude_demodulation(signal, sample_rate):
    # Generate time values based on the length of the input signal and the sample rate
    t = np.arange(len(signal)) / sample_rate
    
    # Generate a carrier signal at a frequency of 18000 Hz
    carrier = np.cos(2 * np.pi * 18000 * t)
    
    # Perform amplitude demodulation by multiplying the signal with the carrier
    demodulated_signal = signal * carrier
    
    return demodulated_signal
```

Lastly, the amplitude_demodulation function is called on Ex3_sound2.wav, Ex3_sound3.wav and Ex3_sound4.wav, and the demodulated signals are displayed using IPython's Audio widget. Upon playing the audios, the secret code "2021" can be heard in Ex3_sound4.wav.

```python
# List of file paths for audio files to be processed
file_paths = ['Ex3_sound2.wav', 'Ex3_sound3.wav', 'Ex3_sound4.wav']

# Iterate through each file path in the list
for file_path in file_paths:
    # Read the WAV file and get sample rate and signal
    sample_rate, signal = scipy.io.wavfile.read(file_path)
    
    # Perform amplitude demodulation on the signal
    demodulated_signal = amplitude_demodulation(signal, sample_rate)
    
    # Print the current file path
    print(file_path)
    
    # Display the demodulated audio using IPython's Audio widget
    display(Audio(demodulated_signal, rate=sample_rate))
    
# Print the suspicious audio file and secret code
print('Suspicious audio file: Ex3_sound4.wav, Secret code: 2021')
```
